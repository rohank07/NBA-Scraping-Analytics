# KNN Clustering NBA Players 

```python
nba_player = pd.read_csv("NBA_player_info_and_stats_2020_2021.csv")
nba_player.drop(columns =["id","height","weight","age","experience"], inplace =True)
nba_player.rename(columns={'Unnamed: 0': 'Name',},inplace= True)
nba_player.set_index('Name', inplace=True)


nba_player.dropna(how="all", axis =0, thresh=4, inplace=True)
nba_player
```

```python
nba_player["PTS"] = nba_player["PTS"].fillna(0)
point_guards = nba_player.loc[nba_player["position"] == "PG"]
point_guards =point_guards[["PTS","AST","STL","REB", "TO", "MIN", "FGA"]]
point_guards.dropna(inplace= True, axis = 0 , thresh=1)
```

```python
point_guards.plot(kind ="scatter", x ="PTS",y ="MIN", color='red')
plt.title("Point Guard Efficiency")
plt.show()
p =point_guards
```

```python
clusters = 4 
points = np.random.choice(point_guards.index, size = clusters)
centroids = point_guards.loc[points]
print(centroids)
fig = plt.figure(num =None, figsize=(6,6), edgecolor='k')
ax = fig.add_subplot(1,1,1, facecolor ="1.0")
ax.scatter(point_guards['PTS'], point_guards['MIN'], alpha =0.5, edgecolor ='g', s=150)
plt.title("Point Guard Efficiency Centroids")
plt.xlabel("Points Per Game")
plt.ylabel("Minutes per Game")
```

```python
def centroids_dict(centroids):
    centroid_dict = dict()
    counter = 0
    for index, row in centroids.iterrows():
        coordinate = [row["PTS"],row["MIN"]]
        centroid_dict[counter] = coordinate
        counter+=1
    return centroid_dict


centroids_dict = centroids_dict(centroids)
    
```

```python
#Eculidean Distance
def eculidean_distance_calculate(centroids, player_stats):
    root_distance = 0
    for x in range(0,len(centroids)):
        difference = centroids[x] - player_stats[x]
        squared_diff = difference **2
        root_distance += squared_diff
    
    euclidean_dist = math.sqrt(root_distance)
    return euclidean_dist

eculidean_distance_calculate([3.5, 10.3], [5.4, 11.5])
```

```python
#Assign stats to clusters

def clusters(player):
    player_val = [player["PTS"], player["MIN"]]
    dist_prev = -1
    cluster_id = None
    
    for centroid_id, centroid_value in centroids_dict.items():
        dist = eculidean_distance_calculate(centroid_value, player_val)
        if dist_prev == -1 or dist <dist_prev:
            cluster_id = centroid_id
            dist_prev = dist
    return cluster_id

point_guards["Cluster"] = point_guards.apply(lambda row: clusters(row), axis = 1)

point_guards
```

```python
point_guards[point_guards["Cluster"] ==2]
```

```python
def visualize_cluster(df, num , iteration):
    colors=['b','y','r','c']
    fig = plt.figure(num=None,figsize=(6,6), edgecolor='k')
    ax = fig.add_subplot(1, 1, 1, facecolor="1.0")
    for i in range(num):
        cluster_df = df[df['Cluster'] == i]
        ax.scatter(cluster_df['PTS'], cluster_df['MIN'],c=colors[i-1], edgecolors='g', alpha=0.5, s=150)
        plt.xlabel("Points Per Game")
        plt.ylabel("Minutes Per Game")
        plt.title('Iteration Number %s'%(iteration))
    plt.show()
iteration =0
visualize_cluster(point_guards, 4, iteration)
```

```python
#sklearn kMeans
point_guards = point_guards[point_guards['PTS'].notna()]
point_guards = point_guards[point_guards['MIN'].notna()]
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters= 4, max_iter=100)
kmeans.fit(point_guards[["PTS","MIN"]])

point_guards['Cluster'] = kmeans.labels_

visualize_cluster(point_guards, 3, 100)
```
