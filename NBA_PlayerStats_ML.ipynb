```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.model_selection as model_selection
import numpy as np
```

```python
stats = pd.read_csv('NBA_player_info_and_stats_2020_2021.csv')
```

```python
stats.drop(columns =["id","height","weight","age","experience"], inplace =True)
```

```python
stats.set_index('Name', inplace=True)
stats.rename(columns={'Unnamed: 0': 'Name',},inplace= True)
```

```python
stats
```

```python
#Predicting the correlation between Average Points scored throughout career and the salary

plt.figure(figsize=(10,10))
stats.plot(kind ='scatter', x="PTS",y="salary", color='blue')
plt.show()
```

```python

points =stats["PTS"].values.reshape(-1,1)
salary = stats["salary"].values.reshape(-1,1)


points =np.nan_to_num(points)
salary=np.nan_to_num(salary)
print(points)
```

```python
points.dtype
salary.dtype
```

```python
from sklearn.linear_model import LinearRegression
x_train,x_test, y_train,y_test = model_selection.train_test_split(points,salary,test_size=0.3,random_state=1)
model = LinearRegression()
model.fit(x_train,y_train)

pred = model.predict(x_test)
rms=np.sqrt(np.mean(np.power((np.array(y_test)-np.array(pred)),2)))
rms
```
